```python
# 01_dataset_preparation.ipynb
# 
# This notebook loads the five NASA PROMISE datasets
# (CM1, JM1, KC1, KC2, and PC1) in CSV format and applies
# the following preprocessing steps to each one:
#
#   1. Detection and reporting of missing values
#   2. Imputation of missing values with zero
#   3. Removal of duplicate instances
#   4. Separation of predictor variables and target variable (defects)
#   5. Removal of non-relevant columns (project, if present)
#   6. Normalization of predictor variables using StandardScaler
#

import pandas as pd
from sklearn.preprocessing import StandardScaler
from pathlib import Path

datasets = [
    {"name": "CM1", "path": "../datasets/CM1-T.csv", "target": "defects"},
    {"name": "JM1", "path": "../datasets/JM1-T.csv", "target": "defects"},
    {"name": "KC1", "path": "../datasets/KC1-T.csv", "target": "defects"},
    {"name": "KC2", "path": "../datasets/KC2-T.csv", "target": "defects"},
    {"name": "PC1", "path": "../datasets/PC1-T.csv", "target": "defects"},
]

processed_path = Path("../datasets/processed")
processed_path.mkdir(parents=True, exist_ok=True)

for ds in datasets:

    print(f"\nProcesando {ds['name']}")

    data = pd.read_csv(ds["path"])

    print(f"Valores faltantes en {ds['name']}:")
    print(data.isnull().sum()[data.isnull().sum() > 0])
    print(f"Total: {data.isnull().sum().sum()}")
    print(f"Porcentaje: {data.isnull().mean().mean()*100:.2f}%\n")


    data = data.fillna(0)
    before_count = len(data)
    data = data.drop_duplicates()
    after_count = len(data)
    print(f"{ds['name']}: {before_count} original instances → {after_count} after removing duplicates ({before_count - after_count} removed)")

    target = ds["target"]

    drop_cols = []
    if "project" in data.columns:
        drop_cols.append("project")

    X = data.drop(columns=[target] + drop_cols)
    y = data[target]

    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    X_scaled = pd.DataFrame(X_scaled, columns=X.columns)

    processed_data = pd.concat([X_scaled, y.reset_index(drop=True)], axis=1)

    output_file = processed_path / f"{ds['name']}_processed.csv"
    processed_data.to_csv(output_file, index=False)
```
